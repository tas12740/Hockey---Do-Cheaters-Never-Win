```python
import pandas as pd
import json
pd.set_option('display.max_columns', 500)
```

```python
file_name = 'hockeyscrape/allgamelogs.json'
```

```python
with open(file_name) as fp:
    hockey_json = json.load(fp)
```

```python
hockey_df = pd.DataFrame(hockey_json)
```

```python
hockey_df.head()
```

```python
hockey_df['game_location'] = hockey_df['game_location'].apply(lambda x: 'A' if x == '@' else 'H')
```

```python
teams = [('ANA', 'MDA'),
         ('ARI', 'PHX', 'WIN'),
         ('BOS',),
         ('BUF',),
         ('CGY', 'ATF'),
         ('CAR', 'HAR'),
         ('CHI', 'CBH'),
         ('COL', 'QBN'),
         ('CBJ',),
         ('DAL', 'MNS'),
         ('DET', 'DTF', 'DTC'),
         ('EDM',),
         ('FLA',),
         ('LAK',),
         ('MIN',),
         ('MTL',),
         ('NSH',),
         ('NJD', 'CLR', 'KCS'),
         ('NYI',),
         ('NYR',),
         ('OTT',),
         ('PHI',),
         ('PIT',),
         ('SJS',),
         ('STL',),
         ('TBL',),
         ('TOR', 'TRS', 'TRA'),
         ('VAN',),
         ('VEG',),
         ('WSH',),
         ('WPG', 'ATL')
        ]

len(teams)
```

```python
for team in teams:
    curr_name = team[0]
    
    df = hockey_df[hockey_df['team'] == curr_name]
    for val in team[1:]:
        df = pd.concat([df, hockey_df[hockey_df['team'] == val]])
        
    df.to_csv(f'teams/{curr_name}_gamelogs.csv', index=False)
```

```python
hockey_df.to_csv('allgamelogs.csv', index=False)
```

```python
team_names = [val[0] for val in teams]

print (team_names)
```
