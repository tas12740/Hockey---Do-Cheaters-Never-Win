```python
import json
```

```python
file_name = 'hockeyscrape/canes.json'
```

```python
with open(file_name) as fp:
    hockey_json = json.load(fp)
```

```python
import pandas as pd
pd.set_option('display.max_columns', 500)
```

```python
hockey_df = pd.DataFrame(hockey_json)
```

```python
hockey_df.head()
```

```python
def convert_game_location(loc):
    if loc == '@':
        return 'A'
    return 'H'
```

```python
hockey_df['game_location'] = hockey_df['game_location'].apply(convert_game_location)
```

```python
hockey_df = hockey_df.fillna(value='')
```

```python
hockey_df.describe()
```

```python
out_name = 'canes.csv'
```

```python
hockey_df.to_csv(out_name,index=False)
```
