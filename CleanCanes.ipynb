```python
import json
```

```python
file_name = 'hockeyscrape/canes.json'
```

```python
with open(file_name) as fp:
    hockey_json = json.load(fp)
```

```python
import pandas as pd
pd.set_option('display.max_columns', 500)

import numpy as np
```

```python
types = {row['curr_type'] for row in hockey_json}
    
print (types)
```

```python
penalty_df = pd.DataFrame([row for row in hockey_json if row['curr_type'] == 'penalty'])
penalty_df.drop(columns=['curr_type'], inplace=True)
```

```python
penalty_df.head()
```

```python
penalty_df.describe()
```

```python
games_df = pd.DataFrame([row for row in hockey_json if row['curr_type'] == 'game_log'])
games_df.drop(columns=['curr_type'], inplace=True)

for col in games_df.columns:
    games_df[col] = games_df[col].apply(lambda val: np.nan if val == '' else val)
```

```python
games_df.head()
```

```python
skaters_df = pd.DataFrame([row for row in hockey_json if row['curr_type'] == 'skaters'])
skaters_df.drop(columns=['curr_type'], inplace=True)

for col in skaters_df.columns:
    skaters_df[col] = skaters_df[col].apply(lambda val: np.nan if val == '' else val)
```

```python
skaters_df.head()
```

```python
skaters_df.describe()
```

```python
goalies_df = pd.DataFrame([row for row in hockey_json if row['curr_type'] == 'goalies'])
goalies_df.drop(columns=['curr_type'], inplace=True)

for col in goalies_df.columns:
    goalies_df[col] = goalies_df[col].apply(lambda val: np.nan if val == '' else val)
```

```python
goalies_df.head()
```

```python
goalies_df.describe()
```

```python
convert_game_location = lambda loc: 'A' if loc == '@' else 'H'
```

```python
games_df['game_location'] = games_df['game_location'].apply(convert_game_location)
```

```python
games_df.describe()
```

```python
games_df
```

```python
out_name = 'canes.csv'
```

```python
games_df.to_csv(out_name, index=False)
```

```python
out_skaters_name = 'canes_skaters.csv'
```

```python
skaters_df.to_csv(out_skaters_name, index=False)
```

```python
out_goalies_name = 'canes_goalies.csv'
```

```python
goalies_df.to_csv(out_goalies_name, index=False)
```

```python
out_penalty_name = 'canes_penalties.csv'
```

```python
penalty_df.to_csv(out_penalty_name, index=False)
```
